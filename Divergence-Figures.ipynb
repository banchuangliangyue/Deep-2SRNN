```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from pylab import rcParams
rcParams['figure.figsize'] = 10, 5

sns.set(color_codes=True)
```

```python
baseDrive = 'data'
workingDirectory = baseDrive+'/'
```

```python
intra_session_cross_entropy = np.load(workingDirectory+'intra-session_cross-entropy.npy')
inter_session_cross_entropy = np.load(workingDirectory+'inter-session_cross-entropy.npy')
inter_subject_cross_entropy = np.load(workingDirectory+'inter-subject_cross-entropy.npy')

inter_session_cross_entropy_999 = np.load(workingDirectory+'inter-session_cross-entropy_999.npy')
inter_subject_cross_entropy_999 = np.load(workingDirectory+'inter-subject_cross-entropy_999.npy')

inter_session_cross_entropy_1000 = np.load(workingDirectory+'inter-session_cross-entropy_1000.npy')
inter_session_cross_entropy_31 = np.load(workingDirectory+'inter-session_cross-entropy_31.npy')

inter_subject_cross_entropy_1000 = np.load(workingDirectory+'inter-subject_cross-entropy_1000.npy')
inter_subject_cross_entropy_31 = np.load(workingDirectory+'inter-subject_cross-entropy_31.npy')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(7, 7), sharex=True)

axes[0].text(1.3, 0.8, r'$\mu_0={:.2f}$'.format(np.mean(intra_session_cross_entropy)), color='r')
axes[0].set_title("Intra-session")
axes[0].hist(intra_session_cross_entropy, 15, density=True)
axes[0].axvline(np.mean(intra_session_cross_entropy), color='r', linestyle='--')

axes[1].text(4, 0.4, r'$\mu_1={:.2f}$'.format(np.mean(inter_session_cross_entropy)), color='r')
axes[1].text(1, 0.4, r'$\mu_1^l={:.2f}$'.format(np.mean(inter_session_cross_entropy_1000)), color='grey')
axes[1].text(1, 0.8, r'$\mu_1^d={:.2f}$'.format(np.mean(inter_session_cross_entropy_31)), color='g')
axes[1].set_title("Inter-session domain shift")
axes[1].hist(inter_session_cross_entropy, 40, density=True)
axes[1].axvline(np.mean(inter_session_cross_entropy), color='r', linestyle='--')
axes[1].axvline(np.mean(inter_session_cross_entropy_1000), color='grey', linestyle=':')
axes[1].axvline(np.mean(inter_session_cross_entropy_31), color='g', linestyle='-')

axes[2].text(6.5, 0.4, r'$\mu_2={:.2f}$'.format(np.mean(inter_subject_cross_entropy)), color='r')
axes[2].text(1, 0.4, r'$\mu_2^l={:.2f}$'.format(np.mean(inter_subject_cross_entropy_1000)), color='grey')
axes[2].text(1, 0.6, r'$\mu_2^d={:.2f}$'.format(np.mean(inter_subject_cross_entropy_31)), color='g')
axes[2].set_title("Inter-subject domain shift")
axes[2].hist(inter_subject_cross_entropy, 80, density=True)
axes[2].axvline(np.mean(inter_subject_cross_entropy), color='r', linestyle='--')
axes[2].axvline(np.mean(inter_subject_cross_entropy_1000), color='grey', linestyle=':')
axes[2].axvline(np.mean(inter_subject_cross_entropy_31), color='g', linestyle='-')

axes[2].set_xlabel("Cross-entropy loss")

plt.xlim(0.0, 16.0)

#fig.text(0.5, 0.04, 'Cross-entropy loss', ha='center', va='center')
#fig.text(0.06, 0.5, 'Histogram', ha='center', va='center', rotation='vertical')

fig.tight_layout()
plt.show()
fig.savefig('divergences_lo-res.png', dpi=300)
fig.savefig('divergences.png', dpi=1200)
```
